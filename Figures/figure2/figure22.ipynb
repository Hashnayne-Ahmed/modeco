```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
cos_phi_r_list = [0.75, 0.80, 0.85, 0.90, 0.95, 1.0]
delta_values = [0.05, 0.07, 0.09, 0.11, 0.13, 0.15]
color_map = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#17becf']
markers = ['o', 's', 'd', '^', 'v', 'p']

# Discretized speed domain
r_hat_vals = np.linspace(0, 1, 500)
marknum = max(1, len(r_hat_vals) // 9)

# Set global font and figure
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.figure(figsize=(10, 6))
legend_lines = []

# Loop over each case
for cos_phi_r, color, marker in zip(cos_phi_r_list, color_map, markers):
    E_over_E0_list = []
    for delta in delta_values:
        f_r_hat = r_hat_vals**3 * (cos_phi_r - delta * (1 - r_hat_vals))
        E_over_E0 = np.trapezoid(f_r_hat, r_hat_vals) / cos_phi_r
        E_over_E0_list.append(E_over_E0)
        plt.plot(
            r_hat_vals, f_r_hat,
            color=color,
            linewidth=1.2,
            marker=marker,
            markersize=4,
            markevery=marknum,
            markerfacecolor='white',
            markeredgewidth=1.2
        )

    # Mean energy line
    E_mean = np.mean(E_over_E0_list)
    plt.axhline(y=E_mean, linestyle='--', linewidth=1.5, color=color, alpha=0.8)

    # Legend entry
    E_min = min(E_over_E0_list)
    E_max = max(E_over_E0_list)
    legend_label = (
        rf'$\cos(\phi_r) = {cos_phi_r:.2f}$, '
        rf'{E_min:.3f}  $\hat{{E}}/\hat{{E}}_0$  {E_max:.3f}'
    )
    legend_lines.append(
        plt.Line2D([], [], color=color, linestyle='-', linewidth=2,
                   marker=marker, markersize=6,
                   markerfacecolor='white', markeredgewidth=1.2,
                   label=legend_label)
    )

# Plot settings
plt.xlabel(r'$\hat{r}(\tau)$', fontsize=32)
plt.ylabel(r'$f(\hat{r}(\tau))$', fontsize=32)
plt.xlim(0, 1)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.6)
plt.legend(handles=legend_lines, fontsize=18, loc='upper left', framealpha=1.0)
plt.tight_layout(pad=0.1)
plt.savefig('figure21.pdf', format='pdf', dpi=1200, bbox_inches='tight')
plt.savefig('figure21.png', format='png', dpi=1200, bbox_inches='tight')
plt.savefig('figure21.eps', format='eps', dpi=1200, bbox_inches='tight')
plt.show()
```
